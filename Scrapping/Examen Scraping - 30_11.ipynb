#### 6. Usar regex para extraer los CUIT del siguiente string

Un CUIT es un código numérico de 10 dígitos.

````python
s = '''Datos del Empleador:Javier Morales \n
CUIT:2039481034 \n Contacto:jmorales@opensource.com 
\n N-ID:813h19djasdj293j \n Datos del Empleador: Rodrigo Perez \n 
CUIT:2384923483 \n Contacto: rperez@opensource.com.ar \n N-ID: 283h1fa93ns8dj3a \n
Datos del Empleador: Sofía Regex \n CUIT:2374813948 Contacto: sregex@amen.py \n N-ID: 29283hdj37g9J2s6'''
````

````python
import re
cuit = "[0-9]{10}"
print(re.findall(cuit,s))
````

#### 7. Hacer un GET a https://www.clarin.com/sitemap.xml y extraer los links usando requests o Selenium, y Beautiful Soup o Regex


````python
import re
import requests
````

````python
def codigo_html(url):
    headers = {'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36'}
    resp = requests.get(url, headers = headers)
    return resp.text
````

````python

clarin = "https://www.clarin.com/sitemap.xml"
regla_clarin = r'<loc>(.+?)<'
links = re.findall(regla_clarin, codigo_html("https://www.clarin.com/sitemap.xml"))
links
````

### 8. Consultar la siguiente API para la lista de valores y guardar los datos en un diccionario

La siguiente URL devuelve chistes de Chuck Norris para la categoría "animales"

https://api.chucknorris.io/jokes/random?category=animal

Con la siguiente lista de categorías posibles, chequear cuáles son válidas viendo el **status_code**, y generar un **diccionario** con un ejemplo **para cada una de las válidas**

````python
categorias = ['animal', 'atypical', 'career', 'celebrity', 'dev', 'distracting', 'explicit',
              'fashion', 'food', 'history', 'money', 'monster', 'movie', 'music',
              'political', 'religion', 'science', 'sport', 'tempting', 'travel']
````

````python
regla_cat = r'value":(.+?)"'
diccionario  = []
for cat in categorias: 
  chistes = re.findall(regla_cat, codigo_html("https://api.chucknorris.io/jokes/random?category=" + cat))
  if len(chistes)>0:
    diccionario.append(cat + " : " + chistes[0])

diccionario
````
